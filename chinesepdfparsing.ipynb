```python
file_path = 'Untitled spreadsheet - Sheet2.csv'
data = pd.read_csv(file_path)
data= data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Function to extract data using regular expressions with adjusted handling for percentage
def regex_extract_adjusted(entry):
    if pd.isna(entry):
        return (None, 0.0)
    match = re.match(r'(\d+-\w+)\s*：\s*.*?(\d+(?:\.\d+)?)%'

, entry)
    if match:
        
        industry_code = match.group(1)  # Extract industry code
        percentage = float(match.group(2) )  # Convert percentage to proportion
        print (percentage+0.0001)
        return (industry_code, percentage)
        
    return (None, 0.0)

# Apply the adjusted regex extraction function
adjusted_regex_data = {}
for city in data.columns:
    city_data = []
    for entry in data[city]:
        code, percent = regex_extract_adjusted(entry)
        print(percent)
        city_data.append((code, percent))
    adjusted_regex_data[city] = city_data

# Refill the DataFrame with adjusted regex extracted data
adjusted_regex_restructured_data = pd.DataFrame(columns=data.columns, index=sorted(set(code for city_list in adjusted_regex_data.values() for code, _ in city_list if code))).fillna(0.0)

for city, entries in adjusted_regex_data.items():
    for code, percent in entries:
        if code:
            adjusted_regex_restructured_data.at[code, city] = float(percent)

adjusted_regex_restructured_data # Display the head of the sorted DataFrame

```
